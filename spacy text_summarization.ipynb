# we have to calcualte the freaquency of each and every word, how many time word is repetation in text

```python
word_frequencies ={}

for word in doc:
    if word.text.lower() not in stopwords:
        if word.text.lower() not in punctuation:
            if word.text not in word_frequencies.keys():
                word_frequencies[word.text] =1
            else:
                word_frequencies[word.text]  +=1
                   
```

```python
word_frequencies
```

```python
len(word_frequencies)
```

```python
max_frequencies =max(word_frequencies.values())
max_frequencies
```

 # to get normalized/weighted frequencies you should devide all frequencies with 11
# weighted toknize---->indivisualtoken/max token

```python

for word in word_frequencies.keys():
    word_frequencies[word] =word_frequencies[word]/max_frequencies
```

```python
#this is the normalized frequencies of each word
#print(word_frequencies)
word_frequencies
```

```python
sentence_tokens =[sent for sent in doc.sents]
sentence_tokens
```

```python
len(sentence_tokens)
```

# # we are going to calculate the sentence score, to calculate the sentence score

```python
sentence_scores = {}

for sent in sentence_tokens:
    for word in sent:
        if word.text.lower() in word_frequencies.keys():
            if sent not in sentence_scores.keys():
                sentence_scores[sent] = word_frequencies[word.text.lower()]
            else:
                sentence_scores[sent] += word_frequencies[word.text.lower()]
                
```

```python
sentence_scores
```

```python
max_sentence_scores =max(sentence_scores.values())
max_sentence_scores
```

```python
len(sentence_scores)
```

# heapq -Sorted out all small number and Large number

```python
# lets say our case study was 30% senetence with maximum scores
from heapq import nlargest
```

```python
select_length =int(len(sentence_tokens)*0.4)
select_length
```

```python
# you have to select maximum 40% sentence out of all sentences
summary = nlargest(select_length,sentence_scores, key = sentence_scores.get)
```
